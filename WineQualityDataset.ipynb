```python
plt.figure()
df["quality"].hist(bins=10)
plt.title("Quality distribution")
plt.xlabel("quality")
plt.ylabel("count")
plt.show()
```

```python
corr = df.corr(numeric_only=True)["quality"].sort_values(ascending=False)
```

```python
corr.head(8)
```

```python
corr.tail(8)
```

```python
top_features = corr.drop("quality").abs().sort_values(ascending=False).head(3).index
top_features
```

```python
for col in top_features:
    plt.figure()
    plt.scatter(df[col],df["quality"],alpha=0.3)
    plt.title(f"{col} vs quality")
    plt.xlabel(col)
    plt.ylabel("quality")
    plt.show()
```
