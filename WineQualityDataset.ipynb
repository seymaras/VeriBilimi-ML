```python
corr = df.corr(numeric_only=True)["quality"].sort_values(ascending=False)
```

```python
corr.head(8)
```

```python
corr.tail(8)
```

```python
top_features = corr.drop("quality").abs().sort_values(ascending=False).head(3).index
top_features
```

```python
for col in top_features:
    plt.figure()
    plt.scatter(df[col],df["quality"],alpha=0.3)
    plt.title(f"{col} vs quality")
    plt.xlabel(col)
    plt.ylabel("quality")
    plt.show()
```

```python
columns=df.columns
(fig,ax)=plt.subplots(4,4, figsize=(16,15))
axes = ax.flatten()

for i,column in enumerate(columns):
    sns.kdeplot(
        data=df,
        x = column,
        hue = "quality",
        ax=axes[i]
    )

    axes[i].set_title(f"{column} Distribution")
    axes[i].set_xlabel(None)

for i in range(i+1, len(axes)):
    axes[i].axis("off")
```
