```python
data = np.array([12, 15, 20, 22, 30, 35, 40])
```

```python
print("Ortalama:", np.mean(data))
print("Medyan", np.median(data))
print("Std:", np.std(data))
print("Max:", np.max(data))
print("Min:", np.min(data))
```

Bulun:
1️⃣ Ortalama maaş
2️⃣ Medyan maaş
3️⃣ En düşük ve en yüksek maaş

```python
salaries = np.array([30000, 45000, 50000, 60000, 75000,
                     80000, 90000, 120000, 150000])
```

```python
print("Ortalama maaş: ", np.mean(salaries))
print("ortalama medyan: ", np.median(salaries))
print("min maaş: ", np.min(salaries))
print("max maaş: ", np.max(salaries))
```

Bulun:
1️⃣ 70.000 TL’den yüksek maaşlar
2️⃣ 50.000 – 100.000 TL arasındaki maaşlar

```python
print(salaries>70000)
```

```python
salaries[salaries>70000]
```

```python
salaries[(salaries>50000)&(salaries<100000)]
```

Yapın:

40.000’den küçük olanları 0 yapın


```python
salaries[salaries<400000]=0
```
