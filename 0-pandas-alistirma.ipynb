Bulun:

3️⃣ En düşük ve en yüksek maaş

```python
salaries = np.array([30000, 45000, 50000, 60000, 75000,
                     80000, 90000, 120000, 150000])
```

```python
print("Ortalama maaş: ", np.mean(salaries))
print("ortalama medyan: ", np.median(salaries))
print("min maaş: ", np.min(salaries))
print("max maaş: ", np.max(salaries))
```

Bulun:

```python
salaries[salaries>70000]
```

#### Zaman var: trend analizi için “yıl bazına indir”

Neyi elde ederiz: yıllara göre katılım, madalya sayısı trendi

```python
df.groupby("Year").size().head()
```

```python
medals_by_year= df.dropna(subset=["Medal"]).groupby("Year").size()
medals_by_year.head()
```

```python
year_total = df.groupby("Year").size()
year_medal = df["Medal"].notna().groupby(df["Year"]).sum()

medal_rate=(year_medal/year_total).sort_index()
medal_rate.head()
```
