```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1= np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
type(my_fig)
type(my_axes)
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1,data2)
my_axes.plot(data2,data1)
plt.show()
```

```python
my_fig, my_axes = plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1,data2,color="#1E0703", alpha=0.5)
my_axes.plot(data2,data1, color="#53CA0A")
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()

new_axes.plot(data1,data1, color="blue", linewidth=1.0)
new_axes.plot(data1,data1+4,color="yellow", linewidth=4.0)

new_axes.plot(data1, data1 +5,color="red", linestyle="-.")
new_axes.plot(data1, data1 +7,color="red", linestyle=":")
new_axes.plot(data1, data1 +9,color="red", linestyle="--")
new_axes.plot(data1, data1 +10,color="red", linestyle="--", marker="o", markersize=8, markerfacecolor="red")
new_axes.plot(data1, data1 +11,color="red", linestyle="-.", marker="+",markersize=4)
plt.show()
```
