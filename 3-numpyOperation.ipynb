#### z-score

```python
data= np.array([10,12,13,15,18,25,100,105])
```

#### outlier

```python
mean=np.mean(data)
```

```python
mean
```

```python
data.sum()/len(data)
```

```python
std=np.std(data)
```

```python
std
```

```python
z_scores=(data-mean)/std
```

```python
z_scores
```

```python
data[z_scores>1]
```
